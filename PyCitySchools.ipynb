```python
# load dependancies
import pandas as pd
import numpy as np
import os
```

```python
# Load files
print(os.getcwd())

schools_load = os.path.join('Resources', 'schools_complete.csv')
students_load = os.path.join('Resources', 'students_complete.csv')

schools_df = pd.read_csv(schools_load, encoding='utf_8')
students_df = pd.read_csv(students_load, encoding='utf_8')

students_df.head()
```

```python
# Create lists to store prefix and suffix
prefixes_suffixes = ["Dr. ", "Mr. ", "Mrs. ", "Miss ", "Ms. ", " DDS", " DVM", " MD", " PhD"]

for word in prefixes_suffixes:
    students_df["student_name"] = students_df["student_name"].str.replace(word, "")
    
students_df.head(20)
```
